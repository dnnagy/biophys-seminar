```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(8,4), dpi=112)
x = np.linspace(-1, 1, 100)
def sigmoid(x):
    return 1/(1+np.exp(-10*x))
  
def theta(x):
    return 1 if x>0 else 0
theta = np.vectorize(theta)

plt.xlabel("time-averaged potential", fontsize=16)
plt.ylabel("firing frequency", fontsize=16)
plt.xticks([])
plt.yticks([])
plt.plot(x, sigmoid(x), ":", label="reality")
plt.plot(x, theta(x), "-", label="mathematical model")
plt.legend()
plt.savefig("sigmoid-model.png", dpi=112)
```
